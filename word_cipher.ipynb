```python
# write to a csv file all the words for a list
def write_words_to_csv():
    words=[]
    for i in range(25):
        word=input("Enter a word: ")
        words.append(word.lower())
    file = open('words.csv', 'w')
    for word in words:
        file.write(word + '\n')
    print("Words written to words.csv")
    file.close()
write_words_to_csv()
```

```python
import numpy as np
import pandas as pd
import itertools
from typing import List

ALPHABET = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']

def main():
    words=open_words_csv()
    word_matrix=pd.DataFrame(words)
    print("Word matrix:\n",word_matrix.to_string(index=False, header=False),"\n\n")
    
    
    # Creates a 2D array of numbers corresponding to each letter in the words
    numbers=[[get_word_numbers(word) for word in words[i]] for i in range(5)]
    
    
    # Get the numeric cores for each set of numbers in the 2D matrix
    cores=[[get_numeric_core(numbers[i][j]) for j in range(5)] for i in range(5)]
    core_matrix=pd.DataFrame(cores)
    print("Core matrix:\n",core_matrix.to_string(index=False, header=False),"\n\n")


# Get the number for each letter in a word 
def get_word_numbers(word):
    return [ALPHABET.index(letter)+1 for letter in word]

# find the numeric core for each set of values in the numbers list
# the goal is to find the lowest possible positive integer possible using +, -, *, /
# There are a total of 16 possible permutations keeping the ordinal values
# this algorithm comes from @Uzarel on github
def get_numeric_core(nums: List[int]):
    
    if len(nums)!=4:
        raise ValueError("List must contain exactly four numbers.")
    if any(n <= 0 for n in nums):
        raise ValueError("All numbers must be positive integers.")
    
    min_core=None
    
    # the first operator is implicitly addition
    for ops in itertools.permutations(['-', '*', '/'], 3):
        core=nums[0]
        div0=False
        for n, op in zip(nums[1:], ops):
            if op=='-':
                core=core-n
            elif op=='*':
                core=core*n
            elif op=='/':
                if n==0:
                    div0=True
                    break
                core=core/n
        #Skip if division by zero
        if div0:
            continue
        if abs(core-round(core))>1e-9:
            continue
        core=int(round(core))
        # ignore non-positive cores
        if core<=0:
            continue
        if min_core is None or core<min_core:
            min_core=core
    return min_core if min_core is not None else None

# open the document containing the words
def open_words_csv():
    file = open('words.csv', 'r')
    words = file.readlines()
    words = [word.strip() for word in words]
    file.close()
    
    return [words[i:i+5] for i in range(0, len(words),5)]
main()
```
